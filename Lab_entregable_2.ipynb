```python
# Parte 1 Escribir un programa que reciba una lista de numeros enteros y determine la cantidad
#De números pares e impares.

# Datos

Lista = list(range(31))

#Imprimir los elementos de la lista

print(Lista)

#  Función Par_impar: Me recibe una lista de enteros
# y me devuelve el conteo de pares e impares

def Par_impar(Lista):
 pares = 0
 impares = 0
# por cada numero en la lista, revisa si el número es par o impar
 for numero in Lista:
  if numero % 2 == 0:
    pares += 1
  else:
    impares +=1
 return pares, impares

pares, impares = Par_impar(Lista)
# Imprime el resultado final del ejercicio

print("Pares:", pares)
print("Impares:", impares)
```
